```python
import pandas as pd
data=pd.read_csv("hire.csv")
data
```

```python
data.iloc[25]#iloc stands for integer location 
#shows the row values at 25th index
```

```python
data.iloc[[25,20]]#shows row values at 25th and 20th index
```

```python
data.iloc[:4]#shows row values from 0 to 3
```

```python
data.iloc[25:]#shows row values from 25 to end
```

```python
data.iloc[4:8]#shows row values from 4  to 7
```

```python
data.iloc[0:10:2]#shows row values from 0 to 9 via increment of index =2
```
