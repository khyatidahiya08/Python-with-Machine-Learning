```python
import pandas as pd#used to read the csv file named hire
data=pd.read_csv("hire.csv")
data
```

```python
data.sort_values("Percentage")
#here we have sorted the values by default in ascending order moreover since we  have the complete table then we need to
#specify that using which column we want to sort the values for that we use by 
#here the value of by is the "Percentage" column
```

```python
data=pd.read_csv("hire.csv",squeeze=True,usecols=["Percentage"])
data#nowe we  squeezed the data to only one column i.e percentage we can sort the values in percentage without having
#to use by
```

```python
data.sort_values().head()#sorts the complete percentage column but shows only first five values (ascending order)
```

```python
data.sort_values(ascending=True).head()#ascending order
```

```python
data.sort_values(ascending=False).head()#descending order
```
