```python
import matplotlib.pyplot as plt
```

# Pie Chart

```python
activities=['eat','sleep','work','play']
slices=[2,7,8,6]#creates list of the slices value i.e what percentage the slices occupy
col=['r','g','m','b']#color of the slices
plt.pie(slices,labels=activities,colors=col,startangle=90,shadow=True)#labels is used to add labels 
#labels and colors are defined attributes
#startangle basically defines the first angle at which the pie chart starts
#shadow is an external boundary shadow on the edges
plt.show()
```

```python
activities=['eat','sleep','work','play']
slices=[2,7,8,6]
col=['r','g','m','b']
plt.pie(slices,labels=activities,colors=col,startangle=90,shadow=True,explode=(0.2,0,0,0),autopct='%1.2f%%')
#explode basically excludes the slice eat out of the pie chart and 0.2 defines the extent of exclusion
#we can exclude other slices also the same way
#autopct stands for auto percentage here %1 stands for the integer part and .2f% stands for two float values 
#the last modulus is for displaying in this format 
#we could display till 3 decimal points by writing .3f%
plt.legend()
plt.show()
```

```python
activities=['eat','sleep','work','play']
slices=[2,7,8,6]
col=['r','g','m','b']
plt.pie(slices,labels=activities,colors=col,startangle=90,shadow=True,explode=(0.2,0,0,0),autopct='%1.1i%%')#here i is used to display only the integer part
plt.legend()#displays the legend box
plt.show()
```
