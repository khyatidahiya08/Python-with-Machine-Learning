```python
import matplotlib.pyplot as plt#imports matplotlib where pyplot is the sub module  and plt is the short name
```

# Line Graph

```python
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('our first plot')
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[2,5,6,4,7,8]
plt.plot(x,y,color='lime')
plt.xlabel('x-axis',color='red')
plt.ylabel('y-axis',color='blue')
plt.title('our second plot',color='green')
plt.show()
```
